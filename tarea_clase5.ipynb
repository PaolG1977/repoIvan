+ Identificador de Números Primos: Crea un programa que solicita al usuario un número entero. Utilizando un bucle para, verificar si el número es primo (es decir, solo divisible por 1 y por sí mismo). Imprime un mensaje indicando si el número es primo o no

```python
numero = int(input("Por favor, ingrese un número entero: "))

primo = True

if numero <= 1:
    primo = False
else:
    for x in range(2, numero):
        if numero % x == 0:
            primo = False
            break

if primo:
    print(f"El número {numero} es primo.")
else:
    print(f"El número {numero} no es primo.")
```

+ Juego de Adivinanza: Genera un número aleatorio entre 1 y 50. Pide al usuario que adivine el número. Utilice un bucle mientras permite múltiples intentos. Da pistas al usuario si el número ingresado es demasiado alto o bajo. Finaliza el juego cuando el usuario adivine correctamente.

```python
import random

numero_maq = random.randint(1, 50)

adivinanza = None

print("Juego de adivinanza de números")
print("Estoy pensando un número entre 1 y 50. ¿Podés adivinar cuál es?")

while adivinanza != numero_maq:
      adivinanza = int(input("Ingresa un número: "))
    
      if adivinanza < numero_maq:
         print("El número es demasiado bajo. Inténtalo de nuevo.")
      elif adivinanza > numero_maq:
           print("El número es demasiado alto. Inténtalo de nuevo.")
      else:
           print(f"¡Felicidades! Adivinaste el número correcto: {numero_maq}")

print("Gracias por jugar!")
```

+ Tabla de Multiplicar: Pide al usuario que ingrese un número entero. Utilice un bucle para imprimir la tabla de multiplicar de ese número del 1 al 10.

```python
numero = int(input("Por favor, ingrese un número entero: "))

print(f"Tabla de multiplicar del {numero}:")

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```
