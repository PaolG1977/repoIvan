5. Imprimir los números del 1 al 100, para los múltiplos de 3, imprimir
“múltiplo de 3”, para los múltiplos de 5 imprimir “múltiplo de 5” y
para los múltiplos de 3 y de 5 imprimir “múltiplo de 3 y 5”.

```python
for nro in range(1, 101):
   
    if nro % 3 == 0 and nro % 5 == 0:
        print(f"El {nro} es múltiplo de 3 y 5")
    
    elif nro % 3 == 0:
        print(f"El {nro} es múltiplo de 3")
    
    elif nro % 5 == 0:
        print(f"El {nro} es múltiplo de 5")
    
    else:
        print(f"El {nro} no es múltiplo de 3 ni de 5")
```

6. Ingresar un número por teclado e imprimir si es número primo o no.


```python
numerop = int(input("Ingrese un número entero: "))

primo = True

if numerop <= 1:
    primo = False
else:
    for x in range(2, numerop):
        if numerop % x == 0:
            primo = False
            break

if primo:
    print(f"El número ingresado: {numerop} es primo.")
else:
    print(f"El número ingresado: {numerop} no es primo.")
```

7. A partir de una lista vacía, utilizar un ciclo while para cargar allí números negativos del -20 al -1

8. A partir de la variable cadena, mostrar en qué posiciones aparece la

programación con Python'


```python
cadena = 'Hola Politécnico. Esto es una práctica de programación con Python'

# Utilizar un bucle for y enumerate para encontrar las posiciones de la letra "n"
for indice, letra in enumerate(cadena):
    if letra == 'n':
        print(f'La letra "n" aparece en la posición {indice}')
```

10. A partir de la siguiente lista de números, crear una lista nueva sólo si
el número es divisible por 5 lista1=[18,25,29,32,35,42,50,60,63,100]


```python
lista1 = [18, 25, 29, 32, 35, 42, 50, 60, 63, 100]

print ("Los números de la lista original son: ", lista1)
# Crear una lista nueva que contenga solo los números divisibles por 5
lista_nueva_div_5 = [numero for numero in lista1 if numero % 5 == 0]

# Imprimir la lista nueva
print("La lista de los números divisibles por 5 son: ", lista_nueva_div_5)
```
